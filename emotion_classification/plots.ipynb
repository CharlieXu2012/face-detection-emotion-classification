```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
train_y = np.load('data/fer2013/train_y.npy')
validation_y = np.load('data/fer2013/validation_y.npy')
test_y = np.load('data/fer2013/test_y.npy')
```

## Data Distributions

```python
plt.bar(['Angry', 'Disgust', 'Fear', 'Happy', 'Sad', 'Suprise', 'Neutral'], np.bincount(train_y))
plt.xlabel('Facial Expressions')
plt.ylabel('Number of Data Points')
plt.title('Training Dataset')
```

```python
plt.bar(['Angry', 'Disgust', 'Fear', 'Happy', 'Sad', 'Suprise', 'Neutral'], np.bincount(validation_y))
plt.xlabel('Facial Expressions')
plt.ylabel('Number of Data Points')
plt.title('Cross Validation Dataset')
```

```python
plt.bar(['Angry', 'Disgust', 'Fear', 'Happy', 'Sad', 'Suprise', 'Neutral'], np.bincount(test_y))
plt.xlabel('Facial Expressions')
plt.ylabel('Number of Data Points')
plt.title('Test Dataset')
```

## VGG 8 - RAW DATA

```python
dir_path = 'data/vgg8_raw/'
loss_history = np.loadtxt(dir_path + 'loss_history.txt')
acc_history = np.loadtxt(dir_path + 'acc_history.txt')
val_loss_history = np.loadtxt(dir_path + 'val_loss_history.txt')
val_acc_history = np.loadtxt(dir_path + 'val_acc_history.txt')
```

```python
n_epochs = 30
plt.plot(range(n_epochs), loss_history)
plt.plot(range(n_epochs), val_loss_history)
plt.xlabel('Number of Epochs')
plt.ylabel('Cross Entropy Loss')
plt.legend(['Training loss', 'Validation loss'])
plt.title('BKVGG8')
```

```python
n_epochs = 30
plt.plot(range(n_epochs), acc_history)
plt.plot(range(n_epochs), val_acc_history)
plt.xlabel('Number of Epochs')
plt.ylabel('Accuracy')
plt.legend(['Training accuracy', 'Validation accuracy'])
plt.title('BKVGG8')
```
